### Chech original text

```python
import json
```

```python
original_text_path = "./dps/datasets/test_korean_jsonl_data/"
```

```python
def list_files(directory):
    files_list = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            files_list.append(os.path.join(root, file))
    return files_list
```

```python
files_list = list_files(original_text_path)
```

```python
original_data = []

for file in files_list:
    with open(file, "r", encoding="utf-8") as file:
        for line in file:
            original_data.append(json.loads(line))
```

```python
for i in original_data[0:10]:
    print(i)
```
